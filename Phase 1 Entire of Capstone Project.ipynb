Firstly we need to install all libraries for our web scraping

```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
from time import sleep
from random import randint
```

```python
headers = {"Accept-Language":"en-US,en;q=0.5"}
```

we need our data in english oonly

```python
Sno = []
Movie_Name = []
Director_Name = []
Duration = []
year = []
ratings = []
Metascore = []
```

need to select all the pages for the web scrap

```python
pages = np.arange(1,1770,50)
```

```python
for page in pages:
    page = requests.get("https://www.imdb.com/search/title/?title_type=feature&num_votes=25000,&genres=action&sort=user_rating,desc&start="+str(page)+"&ref_=adv_nxt")
    soup = BeautifulSoup(page.text,'html.parser')
    movie_data = soup.findAll('div',attrs ={'class':'lister-item mode-advanced'})
    sleep(randint(2,8))
    for store in movie_data:
        sno = store.h3.find('span',class_="lister-item-index unbold text-primary").text
        Sno.append(sno)
        
        name = store.h3.a.text
        Movie_Name.append(name)
        
        director = store.find('p',class_='').find_all('a')[0].text
        Director_Name.append(director)
        
        runtime = store.find("span",class_='runtime').text
        Duration.append(runtime)
        
        Release_year = store.h3.find('span',class_="lister-item-year text-muted unbold").text
        year.append(Release_year)
        
        rate = store.find('div',class_="inline-block ratings-imdb-rating").text.replace('\n','')
        ratings.append(rate)
        
        meta = store.find('span',class_="metascore").text if store.find('span',class_="metascore") else"*****"
        Metascore.append(meta)
        
```

1) so we used Beautifulsoup for our data scraping 

2) in place of all blank columns we fill it up with some characters

```python
print(len(Sno),len(Movie_Name),len(Director_Name),len(Duration ),len(year),len(ratings),len(Metascore))
```

So our Data is equal so we will have acces to make data frame out of it.

```python
Movie_list=pd.DataFrame({"Sno":Sno,"Movie_Name":Movie_Name,"Director_Name":Director_Name,"Duration":Duration,"Ratings":ratings,"Metascore":Metascore})
```

1) So we have total of 1770 rows and 6 columns from our Table 1 dataset

2) Further we need to convert our table into required csv file

```python
Movie_list.to_excel("Movies Table 1 by IMDb.xlsx")
```

```python
Movie_list.to_csv('Movies Table 1 by IMDb.csv')
```

# For Table 2

```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
from time import sleep
from random import randint
```

```python
headers = {"Accept-Language":"en-US,en;q=0.5"}
```

```python
Movie_Name = []
stars = [] 
votes = [] 
Genre = [] 
Gross_collection = [] 
Certification = []
```

```python
pages = np.arange(1,1770,50)
```

```python
for page in pages:
    page = requests.get("https://www.imdb.com/search/title/?title_type=feature&num_votes=25000,&genres=action&sort=user_rating,desc&start="+str(page)+"&ref_=adv_nxt")
    soup = BeautifulSoup(page.text,'html.parser')
    movie_data = soup.findAll('div',attrs ={'class':'lister-item mode-advanced'})
    sleep(randint(2,8))
    for store in movie_data:
        name = store.h3.a.text
        Movie_Name.append(name)
        
        stars.append([a.text for a in store.find('p',class_='').find_all('a')[1:]])
        
        genre = store.find('span',class_="genre").text.strip()
        Genre.append(genre)
        
        certificate = store.find('span',class_='certificate').text if store.find('span',class_='certificate') else '-'
        Certification.append(certificate)
        
        value = store.find_all('span',attrs = {'name':"nv"})
        
        vote = value[0].text
        votes.append(vote)
        
        grosses = value[1].text if len(value)>1 else '%^%^%'
        Gross_collection.append(grosses)
        
    
        
```

```python
print(len(Movie_Name),len(stars),len(votes),len(Genre),len(Gross_collection),len(Certification))
```

```python
Movie_list=pd.DataFrame({"Movie_Name":Movie_Name,"stars":stars,"votes":votes,"Genre":Genre,"Gross_collection":Gross_collection,"Certification":Certification})
```

```python
Movie_list
```

```python
Movie_list.to_excel("Movies Table 2 by IMDb.xlsx")
```

```python
Movie_list.to_csv('Movies Table 2 by IMDb.csv')
```
